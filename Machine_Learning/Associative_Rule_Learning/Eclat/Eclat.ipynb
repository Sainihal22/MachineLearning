```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data Preprocessing
Data = pd.read_csv("Market_Basket_Optimisation.csv", header=None)     # when there is no column names
Data
```

```python
# Need transactions in a list
# Apriori needs string to get trained
transactions = []
for i in range(0,7501):
  transactions.append([str(Data.values[i,j]) for j in range(0,20)])            # Get access to that values of specific location i,j
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
# min support =  transactions per day = 3, in a week = 7*3, total transactions = 7501 min support = 21/7501
# min confidence = by default = 0.8, so try values and choose it
# min lift choosing is based on experience, lowest is 3 and goes upper 3,4,5,6
# min_length and max_length are compulsory arguments which says buy somethings and get something free
# NOT REMOVING LIFT AND CONFIDENCE FOR BETTER ASSOCIATION
```

**Visualizing results**

```python
# Displaying first results directly coming from output of the function
results = list(rules)
results
```

```python
# Putting the results well organised into a Pandas DataFrame
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```
